### EDA ON YIELD DATA

```python
yielddata.describe()
```

```python
yielddata = yielddata[yielddata["year"] >= 1986]
```

```python
crop_avg_yield = yielddata.drop(columns=["year", "country"]).mean().sort_values(ascending=False)

print("\nTop 5 Highest Yielding Crops:")
print(crop_avg_yield.head())

print("\nTop 5 Lowest Yielding Crops:")
print(crop_avg_yield.tail())
```

```python
# Group by 'year' and sum all yield columns
yearly_yield = yielddata.drop(columns=["country"]).groupby("year").sum()

# Find the year with the highest total yield
highest_yield_year = yearly_yield.sum(axis=1).idxmax()
highest_yield_value = yearly_yield.sum(axis=1).max()

# Find the year with the lowest total yield
lowest_yield_year = yearly_yield.sum(axis=1).idxmin()
lowest_yield_value = yearly_yield.sum(axis=1).min()

print(f"Highest Yield Year: {highest_yield_year} with total yield of {highest_yield_value}")
print(f"Lowest Yield Year: {lowest_yield_year} with total yield of {lowest_yield_value}")
```

```python
# Group by 'country' and sum all yield columns
country_yield = yielddata.drop(columns=["year"]).groupby("country").sum()

# Find the country with the highest total yield
highest_yield_country = country_yield.sum(axis=1).idxmax()
highest_yield_value = country_yield.sum(axis=1).max()

# Find the country with the lowest total yield
lowest_yield_country = country_yield.sum(axis=1).idxmin()
lowest_yield_value = country_yield.sum(axis=1).min()

print(f"Highest Yield Country: {highest_yield_country} with total yield of {highest_yield_value}")
print(f"Lowest Yield Country: {lowest_yield_country} with total yield of {lowest_yield_value}")
```

```python
numeric_cols = yielddata.select_dtypes(include=[np.number]).columns
```

```python
numeric_cols = numeric_cols.drop("year", errors="ignore")
```

```python
yielddata[numeric_cols] = yielddata[numeric_cols].apply(lambda x: np.log1p(x)) 
```

```python
yielddata.head(5)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(yielddata[numeric_cols].melt(value_vars=numeric_cols), x="value", bins=30, kde=True)
plt.legend([], [], frameon=False)
plt.xlabel("Transformed Yield")
plt.ylabel("Frequency")
plt.title("Distribution of Log-Transformed Crop Yields")
plt.show()
```

```python
# Compute mean yield per crop
mean_yields = yielddata.drop(columns=["year", "country", "Total_Yield"]).mean().sort_values()

# Lowest 10 crops
lowest_crops = mean_yields.head(10)

# Highest 10 crops
highest_crops = mean_yields.tail(10)

# Plot
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
sns.barplot(x=lowest_crops.values, y=lowest_crops.index, ax=axes[0], palette="Reds_r")
sns.barplot(x=highest_crops.values, y=highest_crops.index, ax=axes[1], palette="Greens_r")

axes[0].set_title("Lowest Yield Crops")
axes[1].set_title("Highest Yield Crops")

plt.tight_layout()
plt.show()
```

## EDA ON YIELD GAP

yield_gap = attainable yield(ideal conditions) -  actual yield(real-world conditions)

```python
yield_gap.describe()
```

- high standard deviation = high inconsistencies
- 0 median suggests attainable~yield
- 25% starts from 1995 indicating recent records

```python
plt.figure(figsize=(8,5))
sns.histplot(yield_gap.drop(columns =["year"]), bins=30, kde=True)
plt.title('Distribution of Yield Gap')
plt.xlabel('Yield Gap')
plt.ylabel('Frequency')
plt.show()
```

```python
attainable.head(5)
```

```python
attainable.tail(5)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=attainable['barley_attainable_yield'], y=yield_gap['barley_yield_gap'])
plt.xlabel("Attainable Yield ")
plt.ylabel("Yield Gap ")
plt.title("Relationship Between Attainable Yield and Yield Gap for Barley")
plt.show()
```

- upward trend indicates attainable yield increases yield_gap also increases.

```python
plt.figure(figsize=(12, 6))
plt.plot(yielddata['year'], yielddata['barley_yield'], label="Actual Yield", marker='o')
plt.plot(attainable['year'], attainable['barley_attainable_yield'], label="Attainable Yield", marker='s')
plt.plot(yield_gap['year'], yield_gap['barley_yield_gap'], label="Yield Gap", marker='x')

plt.xlabel("Year")
plt.ylabel("Yield")
plt.title("Comparison of Barley Yield, Attainable Yield, and Yield Gap Over Time")
plt.legend()
plt.grid(True)
plt.show()
```

- overcrowded means too many observations
- large difference between yield_gap means actual and attainable have large difference
- gap between actual and attainable indicates barley hasn't reached its full potential
